```python
import sqlite3
import pandas as pd
```

```python
connection=sqlite3.connect('escuelita.db')
cursor = connection.cursor()

# Crear la tabla alumnos
cursor.execute("""
CREATE TABLE IF NOT EXISTS alumnos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    apellido_p TEXT,
    apellido_m TEXT,
    fecha_nac TEXT
)
""")
connection.commit()
# Agregar un registro de alumno
cursor.execute(
    "INSERT INTO alumnos (nombre, apellido_p, apellido_m, fecha_nac) VALUES (?, ?, ?, ?)",
    ("Paola", "Vargas", "Aguilar", "2004-12-03")
)
connection.commit()
```

```python
data=pd.read_sql_query('SELECT * FROM alumnos', connection)
```

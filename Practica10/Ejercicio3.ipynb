```python
import matplotlib.pyplot as plt

sport_counts = data['Sport'].value_counts()
total = sport_counts.sum()
labels = []
sizes = []
others_size = 0

for sport, count in sport_counts.items():
    percent = count / total * 100
    if percent < 3:
        others_size += count
    else:
        labels.append(sport)
        sizes.append(count)

if others_size > 0:
    labels.append('Others')
    sizes.append(others_size)

def autopct_func(pct):
    return ('%1.1f%%' % pct) if pct >= 3 else ''

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct=autopct_func, startangle=140)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

np.random.seed(42)

n = 800
date_rng = pd.date_range("2024-01-01", "2024-06-30", freq="D")
regions  = np.array(["North", "South", "East", "West"])
products = np.array(["Widget", "Gadget", "Doohickey"])

df = pd.DataFrame({
    "date": np.random.choice(date_rng, size=n),
    "region": np.random.choice(regions, size=n),
    "product": np.random.choice(products, size=n),
    "units": np.random.poisson(lam=5, size=n) + 1,  # >=1
    "unit_price": np.round(np.random.uniform(10, 100, size=n), 2),
})
df["revenue"] = df["units"] * df["unit_price"]

mask = np.random.rand(n) < 0.8
df["discount"] = np.where(mask, np.round(np.random.uniform(0.0, 0.15, size=n), 3), np.nan)

df = df.sort_values("date").reset_index(drop=True)

display(df.head())
```

```python
# Cell 1: Inspect & select
# Inspect
buf = []
df.info()
display(df.select_dtypes(include="number").describe())

# Select columns
display(df[["date", "region", "revenue"]].head(8))

# Row filtering with .loc (example: North region in January 2024)
jan_north = df.loc[
    (df["region"].eq("North")) &
    (df["date"].between("2024-01-01", "2024-01-31")),
    ["date", "region", "product", "units", "revenue"]
]
display(jan_north.head(8))

# Handling missing values
df["discount_filled"] = df["discount"].fillna(0.0)
df[["discount", "discount_filled"]].head(5)
```

```python
# GroupBy + aggregation
revenue_by_region = (
    df.groupby("region", as_index=False)["revenue"]
      .sum()
      .sort_values("revenue", ascending=False)
)
display(revenue_by_region)

stats_by_product = (
    df.groupby("product")
      .agg(units_mean=("units", "mean"),
           revenue_sum=("revenue", "sum"),
           orders=("product", "count"))
      .reset_index()
      .sort_values("revenue_sum", ascending=False)
)
display(stats_by_product)
```

```python
# Time series resample + pivot
monthly_total = (
    df.set_index("date")["revenue"]
      .resample("M")
      .sum()
      .reset_index()
)
display(monthly_total)

monthly_by_region_long = (
    df.set_index("date")
      .groupby("region")["revenue"]
      .resample("M")
      .sum()
      .reset_index()
)

pivot_month_region = (
    monthly_by_region_long
      .pivot(index="date", columns="region", values="revenue")
      .fillna(0.0)
)
display(pivot_month_region)
```

```python
# Cell 4: Line plot
plt.figure()
plt.plot(monthly_total["date"], monthly_total["revenue"], label="Monthly revenue")
plt.title("Monthly Total Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Scatter + annotation
sample = df.sample(200, random_state=0)

plt.figure()
plt.scatter(sample["units"], sample["revenue"], label="Orders")

# Annotate the point with max revenue
max_idx = sample["revenue"].idxmax()
plt.annotate("Max revenue",
             xy=(sample.loc[max_idx, "units"], sample.loc[max_idx, "revenue"]),
             xytext=(10, 10), textcoords="offset points",
             arrowprops=dict(arrowstyle="->"))

plt.title("Units vs Revenue (sample)")
plt.xlabel("Units")
plt.ylabel("Revenue")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Bar chart
plt.figure()
plt.bar(revenue_by_region["region"], revenue_by_region["revenue"], label="Revenue by region")
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.grid(True, axis="y")
plt.legend()
plt.show()
```
